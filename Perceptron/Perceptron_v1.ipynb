```python
import os,random
import numpy as np
from random import shuffle
```

```python
#Mushroom into training_data and test_data
with open('./data/mushroom.txt') as mushroom:
    a = mushroom.read()
    a = a.split('\n')

total_data=[]
for i in a[:-2]:
    b=i.split(',')
    total_data.append(b)

raw_data = np.array(total_data)
np.random.seed(1)
np.random.shuffle(raw_data)

```

```python
##mushroom feature attribute
#feature={'class':{'edible','poisonous'},'cap-shape':{'bell','conical','convex','flat','knobbed'},'cap-surface..
attributes={}
for i in range(0,len(raw_data[0])):
    features = ['class','cap-shape','cap-surface','cap-color','bruises','odor',
                'gill-attachment','gill-spacing','gill-size','gill-color','stalk-shape','stalk-root',
                'stalk-surface-above','stalk-surface-below','stalk-color-above-r','stalk-color-below-r',
                'veil-type','veil-color','ring-number','ring-type','spore-print-color','population','habitat']
    p=set()

    for j in raw_data:
        p.update(j[i])
    attributes[features[i]]=p
print(attributes)
```

```python
#make_data_into_numbers
List=list(attributes.values())
alphabet_number=[] #각 attribute을 숫자로 나타낸 것 
for i in List:
    Dict={}
    n=0
    for j in list(i):
        Dict[j]=n
        n+=1
    alphabet_number.append(Dict)
```

```python
#make text_data into list of numbers
xx_data=[] # 트레이닝 데이터의 x벡터(22개 모든 feature에 대해 숫자로 나타냄)

for j in raw_data:
    n=0
    tt=[]
    for t in j:
        tt.append(alphabet_number[n][t])
        n+=1
    xx_data.append(tt)
x = np.array([i[1:] for i in np.array(xx_data)])
y = np.array([i[0] for i in np.array(xx_data)])
```

```python
import matplotlib.pyplot as plt
import numpy as np

def perceptron(x,y, w=None, eta = 0.05, epochs=100, random_seed = 10): #epochs is how much would you rotate your function
    X = np.c_[np.ones(len(x)),x] #integrate column vector with x in 2order to make a threshold
    if w is None:
        np.random.seed(random_seed)
        w = np.random.random(X.shape[1]) #initial w
    maxlabel, minlabel = y.max(), y.min()   #y.max() is numpy.int32, it can be altered into np.max(y)
    for i in range(epochs):
        for xi,yi in zip(X,y): #zip method is to combine two lists or two vectors into one
            xi = xi.reshape(w.shape) #in case that xi shape and yi shape do not match
            z = np.dot(xi,w.T)
            yhat = np.where(z>=0,maxlabel,minlabel) #where function: if z>=0: return maxlabel, if not: return minlabel
            delta = eta*(yi-yhat)*xi
            w +=delta
    return(np.round(w,2))
```

```python
%%time
for number in np.arange(10)*0.1:
    if number!= 0:
        num = int(x.shape[0]*round(number,2))
        training_x, test_x = x[:num], x[num:]
        training_y, test_y = y[:num], y[num:]

        weight = perceptron(training_x, training_y)
        print(weight)
        validation = []
        test_x = np.c_[np.ones(len(test_x)),test_x]
        for xi,yi in zip(test_x, test_y):
            z = np.dot(xi,weight.T)
            yhat = np.where(z>=0,1,0)
            if yhat == yi:
                validation.append(1)
            elif yhat != yi:
                validation.append(0)

        percent = round((validation.count(1)/len(validation))*100,2)
        print(str(weight)+'\n'+'트레이닝 데이터 갯수: '+ str(num-1)+'\t'+'확률: '+str(percent)+'\n')
```
